```python
# 애플 실리콘이 탑재된 맥 사용자의 경우
import torch


device = "mps" if torch.backends.mps.is_available() and torch.backends.mps.is_built() else "cpu"
cpu = torch.FloatTensor([1, 2, 3])
mps = torch.rand((1, 1), device=device)
print(device)
print(mps)
```
